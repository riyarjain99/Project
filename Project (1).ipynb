Input the fasta filename. 
Get the sequence.
Use this sequence to send to BLAST, use appropriate expect value [can ask user] to get desired blast result ( you can experiment this first on the web based blast ). 
From the resulting blast report in your program, pick up the top 7 matching sequences[can ask this no.] and put them in a multi-fasta file.
Now this multifasta file can be used as an input to generate phylogenetic tree.

In case that many results have not come you can ask the user a different expect value maybe.. or you can ask that we have only so many results do you want to continue with the phylogenetic tree .. etc.

```python
from Bio import SeqIO
from Bio.Blast import NCBIWWW
from Bio.Blast import NCBIXML
Fname=input("Enter fasta file name with extention") # Get file name from user
E_val=input("Enter expect value") # Get E value input from user
res_num=input("How many top results needed")
c=0
sr = SeqIO.read(Fname, "fasta")
result_handle = NCBIWWW.qblast("blastn", "nt", sr.seq, expect=E_val)
save_file = open("my_blast.xml", "w") # open a file that we will create
save_file.write(result_handle.read()) # get content from result_handle
with open("my_blast.xml") as result_handle:
    blast_record = NCBIXML.read(result_handle)
for alignment in blast_record.alignments:
    for hsp in alignment.hsps:
        #print hsp.sbjct
        line1=">"+alignment.title
        line2=hsp.sbjct
        data=line1+"\n"+line2
        with open("mul.fasta", "w") as file1:
                file1.write(">"+alignment.title)
                file1.write(hsp.sbjct)
                c=c+1
    if(c==res_num):
        break
        



```
